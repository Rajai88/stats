```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0,5,0.01)
y1 = np.sin(2*np.pi*t)
y2 = np.cos(2*np.pi*t)
```

```python
fig, (ax1,ax2) = plt.subplots(2,1)
ax1.plot(t,y1,'b-', label = 'sine')
plt.legend()

ax2.plot(t,y2,'g-*', label = 'cosine')
ax1.legend()
ax2.legend()
bbox_to_anchor=(0.3, 0.3)
```
